```
class KeypadButton:
    def __init__(self, data):
        self.text = None
        self.element = None
        self.data = data

    def __repr__(self):
        return f'KeypadButton(text={self.text},element={self.element})'
```

```
def fetch_balance():
    driver.get('https://www.ing.com.au/securebanking/')
    time.sleep(7)

    client_number_field = driver.find_element_by_id('cifField')
    login_btn = driver.find_element_by_id('login-btn')

    client_number_field.send_keys('Your Client Number Here')

    # Keypad positions are randomised, we have to find which order the butotns are in
    keypad = _keypad()
    # Press all the correct key pad numbers based off what we just calculated
    password = 'Your PIN number here'
    for character in password:
        keypad[character].click()

    login_btn.click()
    time.sleep(7)
    
    balance = driver.find_elements_by_xpath('/html/body/ing-index/div/ing-pages/iron-pages/ing-page-container[2]/section/ing-dashboard/div/iron-pages/section[2]/ing-all-accounts-summary/div/ing-page-block[2]/div/div[2]/div[1]/div[1]/div/div/div[2]/div[1]/div[5]/div/span[2]')
    
    for value in balance:
        print(value.text)
    

def _keypad():
    keypad = driver.find_element_by_id('keypad')
    keypad_buttons = keypad.find_elements_by_xpath('//*[@id="keypad"]/div/div/div')
    buttons = []
    for keypad_button in keypad_buttons:
        data = keypad_button.find_element_by_tag_name('img').get_attribute('src')[22:]
        button = _button_by_data(data, buttons)
        if button is None:
            button = KeypadButton(data=data)
            buttons.append(button)
        button.element = keypad_button
    for button in buttons:
        _set_button_value(button)

    return {key.text: key.element for key in buttons}

def _button_by_data(data: str, buttons: List[KeypadButton]) -> Optional[KeypadButton]:
    for button in buttons:
        if button.data == data:
            return button
    return None

def _set_button_value(button: KeypadButton):
    # the exact value of the png data seems to change between page loads, but the general
    # relative length of the fields stays the same. As long as the fields are relatively the
    # same length then we can guess which number they are
    closest_val = None
    closest_diff = 20
    for k, v in num_pad_btns.items():
        length_diff = abs(len(v) - len(button.data))
        if length_diff < closest_diff:
            closest_val = k
            closest_diff = length_diff

    button.text = closest_val
```
